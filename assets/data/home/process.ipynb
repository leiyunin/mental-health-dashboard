```python
df = pd.read_csv('../mental_trend/line_chart.csv')
```

```python
df['Week'] = pd.to_datetime(df['Week'])
```

```python
# group df by month
df = df.groupby(pd.Grouper(key='Week', freq='M')).mean().reset_index()
```

```python
df['month'] = df['Week'].dt.strftime('%Y-%m')
```
