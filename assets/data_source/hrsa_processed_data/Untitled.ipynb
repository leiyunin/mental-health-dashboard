```python
from geopy.geocoders import Nominatim
import pandas as pd

# Initialize Nominatim API
geolocator = Nominatim(user_agent="geoapiExercises")


# Create DataFrame
df = pd.read_csv()

# Function to geocode address
def geocode_address(row):
    try:
        address = f"{row['Site Address']}, {row['Site City']}, {row['Site State']}, {row['Site Zip Code']}, USA"
        location = geolocator.geocode(address)
        if location:
            return pd.Series([location.latitude, location.longitude], index=['Latitude', 'Longitude'])
    except:
        return pd.Series([None, None], index=['Latitude', 'Longitude'])

# Apply the function to each row
coordinates = df.apply(geocode_address, axis=1)
df = pd.concat([df, coordinates], axis=1)

# Save the DataFrame with coordinates to a new CSV file
df.to_csv("clinical_resources_with_coordinates.csv", index=False)
```
