```python
import pandas as pd
import os
```

```python
path = './depression/state/age'
filenames = os.listdir(path)
merged_df = pd.DataFrame({'state':[]})
general_info_flag = 0
for filename in filenames:
    print(os.path.join(path, filename))
    df = pd.read_csv(os.path.join(path, filename))
    filename_list = filename.split('_')
    if filename_list[1] == 'All Ages':
        filename_list[1] = 'all_ages'
    elif filename_list[1] == 'Over 18':
        filename_list[1] = 'adults'
    elif filename_list[1] == 'Under 18':
        filename_list[1] = 'youth'
    if filename_list[0] == 'All Years':
        filename_list[0] = 'all_years'
    filename_head = filename_list[0].lower() + '-' + filename_list[1] + '-' + filename_list[2].lower()
    if general_info_flag == 0:
        column_names = ["state", #2
                        "latitude", #5
                        "longitude", # 7
                        "2021_state_population", # 9
                        f"{filename_head}-total_scoring_severe_depression", #3
                        f"{filename_head}-total_depression_response", #4
                        f"{filename_head}-percent_at_risk_among_screeners", #12
                        f"{filename_head}-#people_at_risk_per_100K_of_state_population"] #13
        df = df.iloc[:,[2,5,7,9,3,4,12,13]]
        general_info_flag = 1
    else:  
        column_names = ["state", #2
                        f"{filename_head}-total_scoring_severe_depression", #3
                        f"{filename_head}-total_depression_response", #4
                        f"{filename_head}-percent_at_risk_among_screeners", #12
                        f"{filename_head}-#people_at_risk_per_100K_of_state_population"] #13
        df = df.iloc[:,[2,3,4,12,13]]
    df.columns = column_names
    merged_df = pd.merge(merged_df,df,on="state",how='outer')

merged_df.to_csv(f'./result/{filename_list[2].lower()}_age_df.csv')
print(f'./result/{filename_list[2].lower()}_age_df.csv')

```
